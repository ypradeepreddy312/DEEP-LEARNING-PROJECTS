### Convert Dataset and label data into array or dataframe

```python
type(dataset)
```

```python
type(label)
```

```python
ary_dataset=np.array(dataset)
ary_label=np.array(label)
```

```python
type(ary_dataset)
```

```python
x=ary_dataset.copy()
y=ary_label.copy()
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Scaling

```python
x_train.max()
```

```python
x_train_scale=x_train/255
```

```python
x_test_scale=x_test/255
```

####

### Sequential Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Conv2D,MaxPooling2D,Flatten,Dense
```

```python
cnn=Sequential()
```

```python
cnn.add(Input(shape=(130,130,3)))
```

### Convolutional layer

```python
cnn.add(Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),padding="valid",activation="relu"))
```

```python
cnn.add(MaxPooling2D(pool_size=(2,2),padding= "valid"))
```

#### 2nd Convolutional layer

```python
cnn.add(Conv2D(filters=42,kernel_size=(3,3),strides=(1,1),padding="valid",activation="relu"))
```

```python
cnn.add(MaxPooling2D(pool_size=(2,2),padding= "valid"))
```

### Fully connected neaural network

```python
cnn.add(Flatten())
cnn.add(Dense(128,activation="relu"))
cnn.add(Dense(256,activation="relu"))
cnn.add(Dense(512,activation="relu"))
cnn.add(Dense(2,activation="softmax"))
```

```python
cnn.summary()
```

```python
cnn.compile(optimizer="adam",
            loss="sparse_categorical_crossentropy",
            metrics=["accuracy"])
```

```python
cnn.compile(
                    optimizer = "adam",
                    loss = "sparse_categorical_crossentropy",
                    metrics = ["accuracy"]
                    )
```

```python
cnn.fit(x_train_scale,y_train,epochs=5)
```

```python
cnn.evaluate(x_test_scale,y_test)
```
