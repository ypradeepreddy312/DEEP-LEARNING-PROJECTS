NLP PIPELINE FUNCTION CODE (preprocessing)

```python
def nlp_pipeline(data):

    data=data.map(lambda x: re.sub("<.+?>", "", x))                                     #remove html tags    
    data=data.map(lambda x: re.sub("https?://|www.+", "", x))                           #remove urls  
    data=data.map(lambda x: x.lower())                                                  #lowercase  
    data=data.map(lambda x: contractions.fix(x))                                        #expand contractions    
    data=data.map(lambda x: "".join([c for c in x if c not in punctuation]))            #remove punctuation
    data=data.map(lambda x: " ".join([w for w in x.split() if w not in sw_eng]))        #remove stopwords   
    data=data.map(lambda x: " ".join([ lemma.lemmatize(w) for w in x.split()]))         #lemmatization  
    return data                                                                         #return the cleaned data    
```
