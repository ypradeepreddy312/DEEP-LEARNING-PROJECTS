### Sequential Model

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Input, Flatten, Dense
```

```python
seq_model = Sequential()
```

```python
seq_model.add(Input(shape = (28, 28)))
```

```python
seq_model.add(Flatten())
```

```python
seq_model.add( Dense(50, activation = "relu"))
```

```python
seq_model.add( Dense(128, activation = "relu"))
```

```python
seq_model.add( Dense(256, activation = "relu"))
```

```python
seq_model.add( Dense(10, activation = "softmax"))
```

```python
seq_model.summary()
```

```python
seq_model.compile( optimizer= "adam",
                   loss = "sparse_categorical_crossentropy",
                   metrics = ["accuracy"])
```

```python
seq_model.fit(X_train, y_train, epochs = 10)
```

```python
seq_model.evaluate(X_test, y_test)
```
